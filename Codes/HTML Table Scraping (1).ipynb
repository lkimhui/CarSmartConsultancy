```python
# import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests as req
import re
```

```python
# configure BeautifulSoup to read html
url = "https://www.sgcarmart.com/used_cars/info.php?ID=1174653"

html_text = req.get(url)
soup = BeautifulSoup(html_text.content, parser="lxml")
```

```python
# scrape table from
html_table = soup.body.find('table', {'id':'carInfo'})
print(html_table.prettify())
```

```python
# use pandas to read the html table
table = pd.read_html(html_table.prettify())
print(len(table))
```

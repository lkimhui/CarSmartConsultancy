```python
# import packages
import requests
from bs4 import BeautifulSoup
import pickle
```

```python
# Scrapes transcript data from twitter
def url_to_transcript(url):
    data = []
    '''Returns transcript data specifically from twitter url'''
    page = requests.get(url)
    # format of page being processed
    soup = BeautifulSoup(page.text, "lxml")
    # define html tag where most of the text reside - get through container's class
    all_tweets = soup.find('div', {'class': 'mw-content-container'}).find_all('p')
    
    # peek into tweets obtained
    print(all_tweets)
    
    # provide a range of text number to be scraped
    for tweet in all_tweets[:100]:
        data += [tweet.text]
        
    # peek into the search result
    print(url)    
    print(data)
    return data

# URLs of transcripts in scope
urls = ['https://en.wikipedia.org/wiki/Blackpink']
```

```python
# Actually request transcripts (takes a few minutes to run)
transcripts = [url_to_transcript(u) for u in urls]
```

```python
# # Pickle files for later use

# # Make a new directory to hold the text files
!mkdir transcripts

with open("transcripts.txt", "wb") as file:
    pickle.dump(transcripts, file)
```
